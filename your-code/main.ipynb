#### Display the trending Python repositories in GitHub

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
#your code
response = requests.get(url)
print(response.status_code)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
````

````python
user = soup.findAll('h1', attrs={'class':'h3 lh-condensed'})
users = [user.find('span', attrs={'class':'text-normal'}).text.replace('\n','').replace('/','').strip() for user in user]
````

````python
repository = soup.findAll('h1', attrs={'class':'h3 lh-condensed'})
repositories = [repository.find('a').text.split()[2] for repository in repository]
````

````python
trend_repos = [(repository + ' (' + user + ')') for repository, user in zip(repositories, users)]
trend_repos
````

####  20 latest earthquakes info (date, time, latitude, longitude and region name) by the EMSC as a pandas dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://www.emsc-csem.org/Earthquake/'
````

````python
#your code
response = requests.get(url)
print(response.status_code)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
````

````python
earthquakes = soup.find('tbody', attrs={'id':'tbody'}).findAll('tr')
````

````python
earthquake = earthquakes[0]
````

````python
earthquakes_dict = [{'date': earthquake.find('td', attrs={'class':'tabev6'}).find('a').text[:10],
    'time': earthquake.find('td', attrs={'class':'tabev6'}).find('a').text[-10:],
    'latitude': earthquake.findAll('td', attrs={'class':'tabev1'})[0].text[:-1] + ' ' + earthquake.findAll('td', attrs={'class':'tabev2'})[0].text[:1],
    'longitude': earthquake.findAll('td', attrs={'class':'tabev1'})[1].text[:-1] + ' ' + earthquake.findAll('td', attrs={'class':'tabev2'})[1].text[:1],
    'depth': earthquake.find('td', attrs={'class':'tabev3'}).text,
    'mag': earthquake.findAll('td', attrs={'class':'tabev2'})[2].text,
    'region_name': earthquake.find('td', attrs={'class':'tb_region'}).text.title().replace('\xa0','')
} for earthquake in earthquakes]
````

````python
earthquakes_df = pd.DataFrame(earthquakes_dict).head(20)
earthquakes_df
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
````

````python
headers = {
    'Content-type': 'text/html;charset=utf-8',
    'accept-charset': 'utf-8'
}
````

````python
#your code
response = requests.get(url, headers=headers)
print(response.status_code)
html = response.content
soup = BeautifulSoup(html, 'html.parser')
````

````python
languages = soup.findAll('div', attrs={'class':'central-featured-lang'})
names = [name.find('strong').text for name in languages]
numbers = [number.find('small').find('bdi').text.replace('\xa0','') for number in languages]
````

````python
for name, number in zip(names, numbers):
    print(name, 'has', number, 'articles.')
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
````

````python
response = requests.get(url)
print(response.status_code)
html = response.text
soup = BeautifulSoup(html, 'html.parser')
````

````python
table = soup.find('table', attrs={'class':'wikitable sortable'}).findAll('tr')[1:11]
row = table[0]
````

````python
languages_dict = [{
    'Language': row.find('a').text,
    'Speakers (millions)': row.findAll('td')[2].text.replace('\n','')
} for row in table]
````

````python
languages_df = pd.DataFrame(languages_dict).head(10)
languages_df
````
